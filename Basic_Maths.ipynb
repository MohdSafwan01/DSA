Problem Statement: Given an integer N, return the number of digits in N.



```python
#brute force approach
def count_digits(n):
 cnt = 0
 while n>0:
  cnt=cnt+1
  n=n//10

 return cnt
```

```python
count_digits(1234)
```

```python
#optimal approach
import math
def count_digits(n):
  #log10 of any number appears to be a no. of digit lesser than it has
  #example: log10(2345) = 3.37014284
  # so adding 1 to it and returning it as integer will give us the no. of digit

  cnt=int(math.log10(n)+1)
  return cnt
```
