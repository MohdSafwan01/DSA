```python
def patterns(N):
  spaces=2 * (N-1)
  #no. of lines
  for i in range(1,N+1):

    #for numbers of initial side
    for j in range(1,i+1):
      print(j,end=" ")

    #for space between the two set of numbers
    for j in range(1,spaces+1):
      print(" ",end=" ")

    #for numbers of the other side
    for j in range(i,0,-1):
      print(j,end=" ")
    print()
    spaces-=2
```

```python
def patterns(n):
  for i in range(0,n):
    #for space
    for j in range(0,n-i):
      print('*',end=' ')
    #for star
    for k in range(0,2*i+1):
      print(' ',end=' ')
    #for space
    for l in range(0,n-i):
      print('*',end=' ')
    print()
  for i in range(n)[::-1]:#to reverse the for loop
       #for space
    for j in range(0,n-i):
      print('*',end=' ')
    #for star
    for k in range(0,2*i+1):
      print(' ',end=' ')
    #for space
    for l in range(0,n-i):
      print('*',end=' ')
    print()
```

```python
def pattern(n):
  spaces=2*n-2
  for i in range(1,2*n-1):
    stars=i
    if i>n:
      stars=2*n-i
    print('*'*stars,end='')

    print(' '*spaces,end='')

    print('*'*stars,end='')

    if i<n:
      spaces-=2
    else:
      spaces+=2
    print()
```
