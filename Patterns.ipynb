```python
def patterns(n):
  for i in range(0,n):
    #for space
    for j in range(0,n-i-1):
      print(' ',end=' ')
    #for star
    for k in range(0,2*i+1):
      print('*',end=' ')
    #for space
    for l in range(0,n-i-1):
      print(' ',end=' ')
    print()
```
